### Hypothesis Testing using chi-square

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Hypothesis Testing using chi-square

# Membuat tabel kontingensi
table = pd.crosstab(df['Churn'], df['Tenure'])

# Uji hipotesis
stat, p, dof, expected = chi2_contingency(table)

# Interpretasi hasil
alpha = 0.05
if p <= alpha:
    print('Terdapat hubungan antara Churn dan Tenure')
else:
    print('Tidak terdapat hubungan antara Churn dan Tenure')
    
print(p)
```
